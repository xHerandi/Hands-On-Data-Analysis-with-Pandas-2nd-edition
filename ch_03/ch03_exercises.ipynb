1. We want to look at data for the Facebook, Apple, Amazon, Netflix and Google (FAANG) stocks, but we were given each as a separate CSV file (obtained using the `stock_analysis` package we will build in Ch7). Combine them into a single file and store the dataframe of the FAANG data as `faang` for the rest of the exercises:

a) Read in the `aapl.csv, amzn.csv, fb.csv, goog.csv`, and `nflx.csv` files </br>
b) Add a column to each dataframe, called `ticker`, indicating the ticker symbol it’s for (Apple’s is AAPL, for example); this is how you look up a stock. In this case, the filenames happen to be the ticker symbols. </br>
c) Append them together into a single dataframe. </br>
d) Save the results in a CSV file called `faang.csv`

```python
#import libraries
import pandas as pd
```

```python
list = ['fb', 'aapl', 'amzn', 'nflx', 'goog']
faang = pd.DataFrame()

for item in list:
    df = pd.read_csv(f'exercises/{item}.csv')   #read each csv from list
    df.insert(0, 'ticker', item.upper())          #new column with ticker symbol
    faang = faang.append(df)                    #add to df initiated above

#saving to new CSV file
faang.to_csv('faang.csv', index=False)
```

2. With `faang`, use type conversion to cast the values of the `date` column into datetimes and the `volume` column into integers. Then, sort by `date` and `ticker`.

```python
df = pd.read_csv('faang.csv')

df = df.astype({
    'date': 'datetime64[ns]',
    'volume': 'int'
}).sort_values(by=['date', 'ticker'])
df.head()
```

3. Find the seven rows in `faang` with the lowest value for `volume`. 

```python
df.nsmallest(7, 'volume')
```

4. Right now, the data is somewhere between long and wide format. Use `melt()` to make it completely long format. Hint: `date` and `ticker` are our ID variables (they uniquely identify each row). We need to melt the rest so that we don’t have separate columns for `open, high, low, close`, and `volume`. 

```python
melted_df = df.melt(
    id_vars= ['ticker', 'date'],
    value_vars= ['open', 'high', 'low', 'close', 'volume'],
    var_name= 'stock_details'
)
melted_df.head()
```

6. The European Centre for Disease Prevention and Control (ECDC) provides an open dataset on COVID-19 cases called *daily number of new reported cases of COVID-19 by country worldwide*. This dataset is updated daily, but we will use a snapshot that contains data from January 1, 2020 - September 18, 2020. Clean and pivot the data so that it's in wide format: <br>

a) read `covid19_cases.csv` file <br>

```python
covid_df = pd.read_csv('exercises/covid19_cases.csv')
covid_df.tail()
```

b) create a `date` column using the data in the `dateRep` column and the `pd.to_datetime()` function <br>

```python
covid_df['date'] = pd.to_datetime(covid_df['dateRep'])

#checking result
covid_df.info()
```

c) set the `date` column as the index and sort the index <br>

```python
covid_df.set_index('date', inplace=True)

covid_df.head()  #index is sorted, starts with 01/01/2020
```

d) replace all occurrences of `United_States_of_America` and `United_Kingdom` with `USA` and `UK`, respectively. Hint: the `replace()` method can be run on the dataframe as a whole <br>

```python
covid_df.replace({
    'United_States_of_America': 'USA', 
    'United_Kingdom': 'UK'
    }, inplace=True)

#checking results
covid_df['countriesAndTerritories'].value_counts()[:11]
```

e) using the `countriesAndTerritories` column, filter the cleaned COVID-19 case data down to Argentina, Brazil, China, Columbia, India, Italy, Mexico, Peru, Russia, Spain, Turkey, the UK, and the USA <br>

```python
#setting list of desired locations
filters = ['Argentina', 'Brazil', 'China', 'Colombia', 'India', 'Italy', 'Mexico', 'Peru', 'Russia', 'Spain', 'Turkey', 'UK', 'USA']

#creating new dataframe of rows which contain a location from the list
new_df = covid_df.loc[
    covid_df['countriesAndTerritories'].isin(filters)
]

#checking results of new dataframe
new_df['countriesAndTerritories'].sort_values().unique()
```

f) pivot the data so the index contains the dates, the columns contain the country names, and the values are the case counts (`cases` column). Be sure to fill in `NaN` values with 0 <br>

```python
pivoted_df = new_df.pivot(
    columns= 'countriesAndTerritories', 
    values= 'cases'
).fillna(0)

pivoted_df.head()
```
